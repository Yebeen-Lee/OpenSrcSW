```python
import csv
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)

lines = ['1호선','2호선', '3호선', '4호선', '경부선','경인선','경원선','안산선','과천선','분당선',
       '일산선','중앙선','장항선','경인선','경충선','수인선','경강선','5호선','6호선','7호선',
        '8호선','9호선','공항철도 1호선','9호선2~3단계','우이신설선'] # 노선명 (총 25개)
max = [-1] * 3 # 이용자 수 상위3개
min = [999999] * 3 # 이용자 수 하위3개
top3 = ['']*3 # 상위 3개 역명
bottom3 = ['']*3 # 하위 3개 역명

for line in lines:
    print("-------------- {0:s} ---------------".format(line))
    for row in data:
        if row[1] == line: # 각 노선 안에서 비교
            users = 0
            for j in range(4,8,1):
                users += int(row[j].replace(',','')) # 각 row당 이용객 수(유임승차+유임하차+무임승차+무임하차)
            for i in range(0,3): # max 구하기
                if users > max[i]:
                    max[i] = users # max
                    top3[i] = row[3] # 역명
                    break
            for i in range(0,3): # min 구하기
                if users < min[i]:
                    min[i] = users # min
                    bottom3[i] = row[3] # 역명
                    break
    print("이용객이 가장 많은 역 순위")
    for i in range(0,3):
        print("{0:d}. {1:s}({2:d})".format(i+1,top3[i],max[i]))
    print("이용객이 가장 적은 역 순위")
    for i in range(0,3):
        print("{0:d}. {1:s}({2:d})".format(i+1,bottom3[i],min[i]))
```
