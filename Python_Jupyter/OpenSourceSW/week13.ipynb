```python
import csv
import matplotlib.pyplot as plt
f = open("temp.csv", 'r',encoding='cp949')
data = csv.reader(f,delimiter=',')
next(data)
minTemp = [] # 최저기온 리스트
maxTemp = [] # 최고기온 리스트
for row in data:
    date = row[0].split('-')
    Y = date[0]
    M = date[1]
    D = date[2]
    if (row[-1]!= ''): # 최고기온
        if (M=='02' and D=='28'):
            maxTemp.append(float(row[-1]))
    if (row[-2]!=''): # 최저기온
        if (M=='02' and D=='28'):
            minTemp.append(float(row[-2]))
plt.title("Temperature of 2/28 (2010-2020)")
plt.plot(minTemp,'ob-', label='min temperature')
plt.plot(maxTemp,'or-', label='max temperature')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f = open("temp.csv", 'r',encoding='cp949')
data = csv.reader(f,delimiter=',')
next(data)
Jan = [] # 1월의 최고기온 리스트
Aug = [] # 8월의 최고기온 리스트
for row in data:
    M = (row[0].split('-'))[1] # 월 데이터
    if (row[-1]!= ''): # 최고기온
        if (M == '01'):
            Jan.append(float(row[-1]))
        elif (M == '08'):
            Aug.append(float(row[-1]))
plt.title("Max Temperature of Seoul at Jan and Aug (2010-2020)")
plt.hist(Jan, bins=100, color='skyblue', label='Jan')
plt.hist(Aug, bins=100, color='orange', label='Aug')
plt.legend()
plt.show()
```
